```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
```

```python
# read in boundary.csv
boundary = pd.read_csv('boundary.csv')
```

```python
# read ratio2020
df = pd.read_csv('ratio2020.csv')
```

```python
# set color
for i in range(len(df)):
    # find the position j where boundary.Total[j] < df.Total[i] < boundary.Total[j+1]
    j = 0
    while boundary.Total[j] < df.Total[i]:
        j += 1
    # if boundary.upper[j] > df.Ratio[i] > boundary.floor[j], set color to green, else red
    if boundary.upper[j] > df.Ratio[i] > boundary.floor[j]:
        plt.plot(df.Total[i], df.Ratio[i], 'go')
    else:
        plt.plot(df.Total[i], df.Ratio[i], 'ro')
```

```python
# draw line graph for boundary.floor and boundary.upper
l1 = plt.plot(boundary.Total, boundary.floor, 'gray')
l2 = plt.plot(boundary.Total, boundary.upper, 'gray')

# fill the area between the curves, color light yellow
plt.fill_between(boundary.Total, boundary.floor, boundary.upper, color='yellow', alpha=0.2)

# set color
for i in range(len(df)):
    # find the position j where boundary.Total[j] < df.Total[i] < boundary.Total[j+1]
    j = 0
    while boundary.Total[j] < df.Number_total[i]:
        j += 1
        if j >= len(boundary.Total) - 1:
            break

    # if boundary.upper[j] > df.Ratio[i] > boundary.floor[j], set color to green, else red
    if df.Number_total[i] < 100000:
        if boundary.upper[j] > df.Ratio_total[i] > boundary.floor[j]:
            plt.plot(df.Number_total[i], df.Ratio_total[i], 'go')
        else:
            plt.plot(df.Number_total[i], df.Ratio_total[i], 'ro')
            # add text df.District[i] to the point
            plt.text(df.Number_total[i], df.Ratio_total[i], df.District[i])

# set x and y axis limits
plt.xlim(0, 100000)
plt.ylim(50, 200)

plt.show()
```
